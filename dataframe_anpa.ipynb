```python
df.Survived=df.Survived.astype('category')
df.Embarked=df.Embarked.astype('category')
df.info()
```

# Filtering data

```python
df[df.Embarked=='C']
```

```python
# <100, female
df[(df.Fare<100) & (df.Sex=='female')]
```

```python
#Treating Null values
df[df.Cabin.isnull()]
```

```python
df.isnull().sum()
```

```python
#Let's throw Cabin away
df=df.drop(['Cabin'],axis=1)
df.head()
```

```python
#replace all missing ages by median value
df.Age=df.Age.fillna(df.Age.median())
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_plot=df.drop(['Embarked','PassengerId','Name','Parch'],axis=1)
```

```python
df_plot.head()
```

```python
sns.pairplot(df_plot,hue='Survived')
```

```python
sns.distplot(df_plot.Age)
```

```python
#Checking if male died more of female
sns.countplot(data=df_plot,x='Sex',hue='Survived')
```

```python
#Impact of age and sex on survival
sns.boxplot(data=df_plot,x='Age',y='Sex',hue='Survived')
```

```python
sns.jointplot(x='Age',y='Fare',data=df_plot,kind='reg')
```
